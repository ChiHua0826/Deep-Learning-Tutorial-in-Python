```python
#引用tensorflow函式庫
import tensorflow
#引用keras函式庫
import keras
#引用csv函式庫
import csv
#引用numpy函式庫
import numpy
#引用pyplot函式庫
import matplotlib.pyplot as plot

#讀取訓練資料
X = numpy.loadtxt("../資料集/X_斜線案例(交叉熵).csv", delimiter=",")
Y = numpy.loadtxt("../資料集/Y_斜線案例(交叉熵).csv", delimiter=",")
#轉換為2x2矩陣大小
X = X.reshape(X.shape[0], 2, 2, 1).astype('float32')
```

```python
#設定亂數種子
numpy.random.seed(2)

#設定神經網路結構
model = keras.models.Sequential()
model.add(keras.layers.Conv2D(filters = 1, kernel_size = (2, 2), activation = 'linear', input_shape = (2, 2, 1)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(2, activation = 'softmax'))

#設定神經網路學習目標
model.compile(loss='categorical_crossentropy', optimizer = 'sgd', metrics = ['acc']) #損失函式: 交叉熵, 最佳化方法: 梯度下降
```

```python
def show_train_history(train_history, x1, x2):
    plot.plot(train_history.history[x1])
    plot.plot(train_history.history[x2])
    plot.title('Train History')
    plot.ylabel('train')
    plot.xlabel('Epoch')
    plot.legend([x1, x2], loc = 'upper right')
    plot.show()
```

```python
#訓練神經網路
train_history = model.fit(X, Y, epochs = 1000, batch_size = 1) #訓練回合數: 1000, #逐筆修正權重
#顯示訓練過程
show_train_history(train_history, 'loss', 'acc')

#顯示訓練錯誤率
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

#顯示神經網路權重值
print(model.get_weights())
```

```python
#將測試資料代入模型進行預測,並取得預測結果
predictions = model.predict(X)
print(predictions)
```
